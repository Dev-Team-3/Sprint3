```python
# Python program to demonstrate  
# finding 3 highest values in a Products  
from heapq import nlargest 
import pandas as pd
import numpy as np
import pymongo
#Connecting to MongoDB
myclient = pymongo.MongoClient("mongodb://localhost:27017/")   

mydb = myclient["DataTracker"]                       

mycol = mydb["top 3 products"]


# Initial Dictionary 
my_prdct = {'Chipsie':250 , 'Fruit & Veg':8000 , 'Chocolate': 4000, 
           'Dessert':250, 'Cooldrinks': 9500, 'Cupcakes': 6000} 
  
print("Initial Product:") 
print(my_prdct, "\n") 
  
ThreeHighest = nlargest(3, my_prdct, key = my_prdct.get) 
  
print("Products with 3 highest values:") 
print("Keys: Values") 
  
for val in ThreeHighest: 
    print(val, ":", my_prdct.get(val)) 
```
